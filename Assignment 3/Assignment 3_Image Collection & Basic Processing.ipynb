#### Step 2: Remove Bad Files


```python
from PIL import Image
import os

path = r"D:\Machine Learning\Assignment\Assignment 3 - Image Collection & Basic Processing\data\raw"

for folder in os.listdir(path):
    folder_path = os.path.join(path, folder)

    for file in os.listdir(folder_path):
        file_path = os.path.join(folder_path, file)

        try:
            img = Image.open(file_path)
            img.verify()

            # Check empty file
            if os.path.getsize(file_path) == 0:
                os.remove(file_path)
                print("Removed empty file:", file)

        except:
            os.remove(file_path)
            print("Removed corrupt file:", file)

print("Cleaning completed")
```

#### Step 3: Display Sample Images

```python
import os
import random
import matplotlib.pyplot as plt
from PIL import Image

# Dataset path
base_path = r"D:\Machine Learning\Assignment\Assignment 3 - Image Collection & Basic Processing\data\raw"
categories = ["cats", "dogs", "birds"]
plt.figure(figsize=(9, 9))
plot_index = 1

# Loop through categories
for category in categories:
    category_path = os.path.join(base_path, category)
    # Get image files
    images = os.listdir(category_path)
 
    # Select 3 random images
    sample_images = random.sample(images, 3)
    for image_name in sample_images:
        image_path = os.path.join(category_path, image_name)

        # Open image
        img = Image.open(image_path)

        # Plot image
        plt.subplot(3, 3, plot_index)
        plt.imshow(img)
        plt.title(category)
        plt.axis("off")
        plot_index += 1

plt.tight_layout()
plt.show()
```

#### Step 4: Check Image Properties

```python
import os
import random
from PIL import Image

# Dataset path
base_path = r"D:\Machine Learning\Assignment\Assignment 3 - Image Collection & Basic Processing\data\raw"

# Get all image paths
all_images = []
for folder in os.listdir(base_path):
    folder_path = os.path.join(base_path, folder)
    for file in os.listdir(folder_path):
        all_images.append(os.path.join(folder_path, file))

# Select 5 random images
sample_images = random.sample(all_images, 5)

# Print image details
for image_path in sample_images:
    img = Image.open(image_path)
    width, height = img.size

    # Number of channels
    channels = len(img.getbands())
    print(f"\nImage: {os.path.basename(image_path)}")
    print("Width :", width)
    print("Height:", height)
    print("Channels:", channels)
```

#### Step 5: Check Image Type

```python
import os
from PIL import Image

# Dataset path
base_path = r"D:\Machine Learning\Assignment\Assignment 3 - Image Collection & Basic Processing\data\raw"

# Counters
rgb_count = 0
grayscale_count = 0
rgba_count = 0

# Loop through folders
for folder in os.listdir(base_path):
    folder_path = os.path.join(base_path, folder)
    for file in os.listdir(folder_path):
        file_path = os.path.join(folder_path, file)
        try:
            img = Image.open(file_path)
            # Check image mode
            if img.mode == "RGB":
                rgb_count += 1
            elif img.mode == "L":
                grayscale_count += 1
            elif img.mode == "RGBA":
                rgba_count += 1
        except:
            pass
# Print results
print("RGB Images      :", rgb_count)
print("Grayscale Images:", grayscale_count)
print("RGBA Images     :", rgba_count)
```
