### ● What is Machine Learning (ML)?

- **Machine Learning (ML) is a branch of artificial intelligence (AI)** where computers learn patterns from data and make decisions or predictions without being explicitly programmed for every task.

- In traditional programming, humans write exact rules for the computer to follow. In machine learning, the process is reversed:

    - You give the computer data + outcomes (examples)
    - The system learns the patterns or relationships
    - Then it uses what it learned to handle new, unseen data

- A simple example:
- If you show a model thousands of emails labeled “spam” or “not spam,” it can learn to identify spam emails on its own later.

- **Machine Learning includes several common approaches:**

    - Supervised learning: learns from labeled data (e.g., predicting house prices)
    - Unsupervised learning: finds hidden patterns (e.g., customer grouping)
    - Reinforcement learning: learns by trial and error using rewards (e.g., game-playing AI)

ML is used in things like recommendation systems (Netflix/YouTube), voice assistants, fraud detection, and self-driving technology.

### ● What is Deep Learning (DL)?

- **Deep Learning (DL)** is a specialized branch of machine learning that uses multi-layered neural networks to learn patterns from large amounts of data.

- It is inspired by how the human brain processes information, using structures called artificial neural networks with many layers (“deep” networks).

- Simple idea:
    - Machine Learning learns from data using algorithms that often need manual feature selection.
    - Deep Learning automatically learns features and patterns by itself using many layers of neural networks.

- Example:

- If you give a deep learning model thousands of cat images:

    - First layers detect edges and shapes
    - Middle layers detect parts like eyes or ears
    - Final layers recognize the full object as a cat
- Where Deep Learning is used:
    - Face recognition (like phone unlock)
    - Speech recognition (voice assistants)
    - Self-driving cars
    - Language translation (like Google Translate)

```python
from sklearn.pipeline import make_pipeline
from sklearn.neighbors import KNeighborsClassifier
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
# transformer =columnTransformer(transformers = [("t1",OrdinalScaler())])

transformer=ColumnTransformer(transformers=[("t1",OneHotEncoder(),[0]),          # categorical
                                            ("t2",StandardScaler(),[1,2,3,4])])    # numerical

# knn_model = make_pipeline(transformer,kNeighborsRegressor())
knn_model = make_pipeline(transformer,KNeighborsClassifier())
```

```python
# create a model
knn_model.fit(X_train ,y_train)
```

```python
# Lets make Prediction
y_pred = knn_model.predict(X_test)
```

```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.metrics import roc_curve, roc_auc_score
from sklearn.preprocessing import label_binarize

# Get unique classes
classes = np.unique(y_test)

# Convert target to binary format
y_test_bin = label_binarize(y_test, classes=classes)

# Get predicted probabilities from pipeline
y_prob = knn_model.predict_proba(X_test)

# Plot ROC for each class
plt.figure(figsize=(8,6))

for i, class_name in enumerate(classes):
    fpr, tpr, _ = roc_curve(y_test_bin[:, i], y_prob[:, i])
    auc = roc_auc_score(y_test_bin[:, i], y_prob[:, i])

    plt.plot(fpr, tpr, label=f"{class_name} (AUC = {auc:.2f})")

# Random line
plt.plot([0,1],[0,1],'k--')

plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("Multi-Class ROC Curve")
plt.legend()
plt.show()
```
